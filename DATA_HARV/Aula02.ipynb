```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
# import matplotlib.pyplot as plt
import requests
```

```python
HTML = requests.get('https://quotes.toscrape.com/')
```

```python
bs = BeautifulSoup(HTML.text, 'lxml')
```

```python
bs
```

```python
bs.find('span')
```

```python
bs.find('span').text
```

```python
quotes = bs.find_all('div', class_='quote')

quotes
```

```python
textos = bs.find_all('span', class_='text')

textos_text = [texto.text for texto in textos]

textos_text
```

```python
autores = bs.find_all('small', class_='author')

autores_text = [autor.text for autor in autores]

autores_text
```

```python
tags = bs.find_all('a', class_='tag')

tags_text = [tag.text for tag in tags]

tags_text
```

```python
df = pd.DataFrame(columns=['autor', 'tags', 'texto'])

for q in quotes:
    autor = q.find('small', class_='author').text
    tags = [tag.text for tag in q.find_all('a', class_='tag')]
    texto = q.find('span', class_='text').text

    df = pd.concat([df, pd.DataFrame({'autor': [autor], 'tags': [tags], 'texto': [texto]})], ignore_index=True)

df['texto'] = df['texto'].apply(lambda x: x.replace('“', '').replace('”', ''))

df
```
